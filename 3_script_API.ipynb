# Extraction des informations concernant les 10 premiers produits

```python
url = "https://edamam-food-and-grocery-database.p.rapidapi.com/api/food-database/v2/parser"

querystring = {"ingr":"\"champagne"}

headers = {
	"X-RapidAPI-Key": "d9e5a99e57msh2c0ac6e76aac869p12cd5ajsn50535546af72",
	"X-RapidAPI-Host": "edamam-food-and-grocery-database.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)
```

```python
json_content = response.json()
for key, value in json_content.items():
    print(key, ":", value)
```

```python
print(json_content["hints"])
```

```python
# by default it returns 20 items for each request (cf. rapidapi website)
nb_produit = len(json_content["hints"])
print(f"Il y a {nb_produit} produits différents à base de champagne")
```

```python
# exemple du 20ème produit
print(json_content["hints"][19]['food'])
```

```python
print("Liste des produits (request n1):")
print("")
for i in range(len(json_content["hints"])):
    print(json_content["hints"][i]['food']['label'])
```

```python
df = pd.DataFrame.from_dict(json_content["hints"])
df = pd.json_normalize(df["food"])
df.head()
```

```python
# on récupère uniquement les data souhaitées
df_top10 = df[["foodId","label","category","foodContentsLabel","image"]]
df_top10 = df_top10.iloc[:10] # il nous est demandé de communiquer les 10 premiers produits
df_top10
```

```python
# extraction des données dans un fichier csv
df_top10.to_csv("champagne_product_first_10.csv")
```

# Test d'extraction des entrées suivantes

```python
# extraction de 200 entrées relatives au champagne (soit 10 requests de 20 entrées chacune)

df_concat = df
json_content_base = json_content

for i in range(2,11):
    url_nxt = json_content_base["_links"]["next"]["href"]
    response_nxt = requests.get(url_nxt)
    json_content_nxt = response_nxt.json()
    df_nxt = pd.DataFrame.from_dict(json_content_nxt["hints"])
    df_nxt = pd.json_normalize(df_nxt["food"])
    df_nxt = df_nxt[["foodId","label","category","foodContentsLabel","image"]]
    df_concat = pd.concat([df_concat, df_nxt], axis=0, ignore_index=True)
    json_content_base = json_content_nxt
```

```python
df_concat = df_concat[["foodId","label","category","foodContentsLabel","image"]]
df_concat
```

```python
df_concat.isna().sum()
```

```python
nb_total = df_concat.shape[0]
nb_image  = df_concat.shape[0] - df_concat["image"].isna().sum()
```

```python
print(f"Sur les {nb_total} aliments extraits, seuls {nb_image} avaient une photo (soit {nb_image/nb_total*100:.2f}%).")
```
